```python
def mostrar_ropa(ropa, tamaño, color):
    print(f"El tipo de ropa que queres es {ropa}, del tamaño de {tamaño} y de color {color}")

def comprar_ropa():
    ropa = input("Ingrese el tipo de ropa a comprar: ")
    tamaño = input("Ingrese el tamaño que quiere: ")
    color = input("Ingrese el color: ")

    mostrar_ropa(ropa, tamaño, color)

comprar_ropa()
```
