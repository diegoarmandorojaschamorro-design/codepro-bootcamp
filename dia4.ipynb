```python
class Superheroe:
    #Atributo de clase
    especie = "superhumanos"
    #Definir atributos de instancia
    def __init__(self, nombre):
        self.nombre = nombre

    #Metodo
    def rayos_laser(self):
        print("tiro rayos laseres")

    def volar(self, km):
        print(f"Estoy volando a {km} km por hora")

    def alzar_objetos(self, objeto, kg):
        print(f"Puedo alzar un {objeto} de {kg} kg")

#Intanciar objetos
superheroe1 = Superheroe("Batman")
superheroe1.volar(120)
```

```python
class Coche:
    tipo = "vehiculo"
    def __init__(self, modelo, posicion = 0):
        self.modelo = modelo
        self.posicion = posicion
        def acelerar(self, metros):
            self.posicion += metros
            print(f"El vehiculo modelo {self.modelo} acelera y avanza {metros}. Nueva posicion : {self.posicion}")
            def girar(self, direccion):
                print(f"El {self.modelo} avanza y gira a la {direccion} direccion")

coche1 = Coche("Toyota")
coche1.acelerar(1)
```

```python
class Mascota:
    especie = "gatuna"

    def __init__(self, nombre, edad, posicion = 0):
        self.nombre = nombre
        self.edad = edad

    def hablo(self, raza):
        print(f"Cuando hablo digo mi{raza}")

    def muevo(self, km):
        self.posicion += km
        print(f"Cuando me muevo voy a {km} por hora")

mascota1 = Mascota("Sam",4,5)
mascota1.hablo{"bills"}
```

```python
class AnimalDeGranja:
    #atributo clase
    tipo = "animal de granja"

    #atributo de instancia
    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    #metodos(acciones)
    def hablar(self):
        print(f"Hablo y digo {self.sonido}")

    def presentarse(self):
        print(f"Hola soy {self.nombre}")


class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevos(self):
        print(f"Estoy poniendo {self.huevos_por_dia} huevos por dia")


class Vaca(Gallina):
    patas = 4

    def mujir(self):
        print("Muu")

gallina1 = Gallina("Hortencia", "marron", "corococo")
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca("Lola", "blanco y negro", "Muuuu")
vaca.hablar()
vaca.mujir()
vaca.patas
```

```python
class TiposDeAutos:
    tipo = "4x4"

    def __init__(self, nombre, color, tipo):
        self.nombre = nombre
        self.color = color
        self.tipo = tipo

    def color_auto(self):
        print(f"El auto es de color {self.color}")

    def nombre_auto(self):
        print(f"El modelo de este auto se llama {self.nombre}")


class Toyota(TiposDeAutos):
    km = 180

    def acelerar(self):
        print(f"Este modelo puede ir a {self.km} por hora")


class Nissan(TiposDeAutos):
    puertas = 4

    def auto(self):
        print(f"Este auto tiene {self.puertas} puertas")

toyota1 = Toyota("Hilux", "blanco", "4x4")
toyota1.color_auto()
toyota1.acelerar()
nissan1 = Nissan("Frontier", "negro", "normal")
nissan1.nombre_auto()
nissan1.auto()
```
