```python
numero_secreto = 7
intentos = 5
print(f"Empezas con {intentos} intentos para encontrar el numero secreto")

while intentos > 0:
    numero = int(input(f"Ingresa un numero entre el 1 y 10: "))
    
    if numero == numero_secreto:
        print(f"Lo lograste. El {numero} es el numero secreto")
        break
    else:
        intentos -= 1
        if intentos > 0:
            if numero_secreto > numero:
                print("El numero secreto es mayor")
            else:
                print("El numero secreto es menor")
            print(f"Te quedan {intentos} intentos")
        else:
            print(f"Fallaste. El numero secreto era el {numero_secreto}")
```

```python
mensaje = input("Ingresa un mensaje:")
cantidad = int(input("Cantidad a repetir:"))

def repetir(m, c):
    for i in range(c):
        print(m)

repetir(mensaje, cantidad)
```

```python
def obt_num():
    numero = int(input("Ingrese un numero a multiplicar:"))
    return numero

def calcular(n):
    resultado = n * 2
    return resultado

numero_obt = obt_num()

resultado_obt = calcular(numero_obt)

print(f"El numero ingresado es: {numero_obt}")
print(f"El resultado de la multiplicacion es: {resultado_obt}")
```
