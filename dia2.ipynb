```python
adivina = 20
ingresa = 10

while ingresa != adivina:
    ingresa = int(input("Ingresa un numero"))
    print(ingresa)
```

```python
numero = 30

while True:
    ingresa = int(input("ingrese un nro"))
    if ingresa == numero:
        print ("el nro es correcto")
        break
    else:
        print("el nro es incorrecto")
```

```python
contraseña = int(input("Crea la contraseña:"))

while True:
    ingresacontra = int(input("Ingrese la contraseña"))
    if ingresacontra == contraseña:
        print("Lo lograste amigo")
        break
    else:
        print("Contraseña incorrecta")
```

lista 

```python
lista_frutas = ["naranja", "pera", "manzana"]
print(lista_frutas)
lista_bool = [True, False]
print(lista_bool)
lista_num = [1,2,3,4] 
print(lista_num)
lista_etc = ["zapato", True, 5.5]
print(lista_etc)
# agarrar 1
print(lista_frutas[1])

#metodo append()
lista_frutas.append("Banana")
print(lista_frutas)

# modificar
lista_frutas[2] = "piñas"
print(lista_frutas)

#sacar elemento de la lista
lista_frutas.remove("naranja")
print(lista_frutas)

# sacar por posicion
lista_frutas.pop(0)
print(lista_frutas)

# cantidad de elementos de una lista
print(len(lista_frutas))

cant_elemen = len(lista_frutas)
print(cant_elemen)
```
